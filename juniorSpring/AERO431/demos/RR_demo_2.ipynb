```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
import ipywidgets
```

```python
x = sym.symbols('x')

phi = []
d2phi = []
for i in range(5):
    phi_i = x**(i + 1) * (1 - x)
    d_phi_i = sym.diff(phi_i, x)
    d2_phi_i = sym.diff(d_phi_i, x)
    phi.append(phi_i)
    d2phi.append(d2_phi_i)

xs_1 = np.linspace(0, 0.75, 100)
xs_2 = np.linspace(0.75, 1, 100)

ws_exact_1 = -xs_1*(15 - 16*xs_1**2)/32
ws_exact_2 = -3*(-9 + 41*xs_2 - 48*xs_2**2 + 16*xs_2**3)/32

def solve_beam_RR(n_basis=2):
    K = np.zeros((n_basis, n_basis))
    F = np.zeros(n_basis)

    for i in range(n_basis):
        K[i,i] = sym.integrate(d2phi[i]**2 / 12, (x, 0, 1))
        for j in range((i + 1), n_basis):
            K[i, j] = sym.integrate(d2phi[i]*d2phi[j]/12, (x, 0, 1))
            K[j, i] = K[i, j]

    for i in range(n_basis):
        F[i] = -phi[i].subs(x, 3/4)

    c = np.linalg.solve(K, F)
    
    w = c[0] * phi[0]
    for i in range(1, n_basis):
        w += c[i] * phi[i]

    xs = np.linspace(0, 1, 100)
    ws = np.array([
        w.subs(x, x_).evalf() for x_ in xs
    ])

    plt.plot(xs, ws, 'k-', lw=3, label='RR')
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.1)
    plt.legend()

ipywidgets.interact(solve_beam_RR, n_basis=(1,5,1))
```
