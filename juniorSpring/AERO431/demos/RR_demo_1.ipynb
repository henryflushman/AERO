```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

```python
xs = np.linspace(0, 1, 100)

xs_1 = np.linspace(0, 0.75, 100)
xs_2 = np.linspace(0.75, 1, 100)

ws_exact_1 = -xs_1*(15 - 16*xs_1**2)/32
ws_exact_2 = -3*(-9 + 41*xs_2 - 48*xs_2**2 + 16*xs_2**3)/32
```

## 1. One-basis approximation.

```python
phi_1 = lambda x: x*(1 - x)

def plot_w1(c1=0.0):
    ws = c1*phi_1(xs)
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='1-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w1, c1=(-1, 1, 0.1))
```

## Two-basis approximation

```python
phi_2 = lambda x: x*x*(1 - x)

def plot_w2(c1=0.0, c2=0.0):
    ws = c1*phi_1(xs) + c2*phi_2(xs)
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='2-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w2, c1=(-1, 1, 0.1), c2=(-1, 1, 0.1))
```

## Three-basis approximation

```python
phi_3 = lambda x: x*x*x*(1 - x)

def plot_w3(c1=0.0, c2=0.0, c3=0.0):
    ws = c1*phi_1(xs) + c2*phi_2(xs) + c3*phi_3(xs)
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='3-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w3, c1=(-1, 1, 0.1), c2=(-1, 1, 0.1), c3=(-1, 1, 0.1))
```
