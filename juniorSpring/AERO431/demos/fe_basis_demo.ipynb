```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

```python
xs = np.linspace(0, 1, 100)

xs_1 = np.linspace(0, 0.75, 100)
xs_2 = np.linspace(0.75, 1, 100)

ws_exact_1 = -xs_1*(15 - 16*xs_1**2)/32
ws_exact_2 = -3*(-9 + 41*xs_2 - 48*xs_2**2 + 16*xs_2**3)/32
```

## One-basis approximation

```python
def phi1_1(x):
    h = 1.0/2
    if x < h:
        return x/h
    else:
        return (1 - x)/h

def plot_w1(c1=0.0):
    ws = np.array([c1*phi1_1(x) for x in xs])
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='1-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w1, c1=(-0.25, 0.25, 0.01))
```

## Two-basis approximation

```python
def phi2_1(x):
    h = 1.0/3
    if x < h:
        return x/h
    elif x < 2*h:
        return (2*h - x)/h
    else:
        return 0.0

def phi2_2(x):
    h = 1.0/3
    if x < h:
        return 0.0
    elif x < 2*h:
        return (x - h)/h
    else:
        return (1 - x)/h 

def plot_w2(c1=0.0, c2=0.0):
    ws = np.array([c1*phi2_1(x) + c2*phi2_2(x) for x in xs])
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='2-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w2, c1=(-0.25, 0.25, 0.01), c2=(-0.25, 0.25, 0.01))
```

## Three-basis approximation

```python
def phi3_1(x):
    h = 1.0/4
    if x < h:
        return x/h
    elif x < 2*h:
        return (2*h - x)/h
    else:
        return 0.0

def phi3_2(x):
    h = 1.0/4
    if x < h:
        return 0.0
    elif x < 2*h:
        return (x - h)/h
    elif x < 3*h:
        return (3*h - x)/h 
    else:
        return 0.0
    
def phi3_3(x):
    h = 1.0/4
    if x < 2*h:
        return 0.0
    elif x < 3*h:
        return (x - 2*h)/h
    else:
        return (1 - x)/h

def plot_w2(c1=0.0, c2=0.0, c3=0.0):
    ws = np.array([c1*phi3_1(x) + c2*phi3_2(x) + c3*phi3_3(x) for x in xs])
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='2-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w2, c1=(-0.25, 0.25, 0.01), c2=(-0.25, 0.25, 0.01), c3=(-0.25, 0.25, 0.01))
```

## Four-basis approximation

```python
def phi4_1(x):
    h = 1.0/5
    if x < h:
        return x/h
    elif x < 2*h:
        return (2*h - x)/h
    else:
        return 0.0

def phi4_2(x):
    h = 1.0/5
    if x < h:
        return 0.0
    elif x < 2*h:
        return (x - h)/h
    elif x < 3*h:
        return (3*h - x)/h 
    else:
        return 0.0
    
def phi4_3(x):
    h = 1.0/5
    if x < 2*h:
        return 0.0
    elif x < 3*h:
        return (x - 2*h)/h
    elif x < 4*h:
        return (4*h - x)/h 
    else:
        return 0.0
    
def phi4_4(x):
    h = 1.0/5
    if x < 3*h:
        return 0.0
    elif x < 4*h:
        return (x - 3*h)/h
    else:
        return (1 - x)/h

def plot_w2(c1=0.0, c2=0.0, c3=0.0, c4=0.0):
    ws = np.array([c1*phi4_1(x) + c2*phi4_2(x) + c3*phi4_3(x) + c4*phi4_4(x) for x in xs])
    plt.plot(xs_1, ws_exact_1, 'k--', lw=2, label='Exact')
    plt.plot(xs_2, ws_exact_2, 'k--', lw=2)
    plt.plot(xs, ws, 'k-', lw=3, label='2-basis')
    plt.xlabel('x')
    plt.ylabel('w')
    plt.xlim(0.0, 1.0)
    plt.ylim(-0.25, 0.25)
    plt.legend()

ipywidgets.interact(plot_w2, c1=(-0.25, 0.25, 0.01), c2=(-0.25, 0.25, 0.01), c3=(-0.25, 0.25, 0.01), c4=(-0.25, 0.25, 0.01))
```
