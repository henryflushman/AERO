```python
import numpy as np
import matplotlib.pyplot as plt
```

Problem 1

```python
# Use Matlab or Python to write code that plots a 500 Mhz square wave with
# the following 3 duty cycles: 25%, 50%, and 75%. Show at least four cycles.
# Show all three waves on the same plot. Label the axes and include a legend

# constants


f = 500e6
T = 1 / f
cyc = 4
tf = cyc * T
dt = T / 20
t = np.arange(0, tf, dt)
duties = [0.25, 0.5, 0.75]
phase = (t % T) / T
waves = [(phase < d).astype(float) for d in duties]

# analysis
plt.figure()
labels = ['25%', '50%', '75%']
for y, lbl in zip(waves, labels):
    plt.plot(t*1e9, y, drawstyle='steps-post', label=lbl)

plt.xlabel('Time (ns)')
plt.ylabel('Amplitude')
plt.title('500 Mhz Square Waves')
plt.legend()
plt.ylim(-.2,1.2)
plt.tight_layout()
plt.show()
```

Problem 2

```python
# constants
k = 1
vc = 5
tau = [100, 50]
t = np.arange(0, 500, .1)

# equation
w = [k*vc*(1-np.exp(-t/i)) for i in tau]

# analysis
plt.figure()
for i in range(len(w)):
    plt.plot(t, w[i], label=f'k = {k}, vc = {vc}, tau = {tau[i]}')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
