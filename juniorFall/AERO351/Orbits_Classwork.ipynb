Hohmann Transfer

```python
mu = 398600
r_earth = 6378
r1 = 300+r_earth
r2 = 3000+r_earth
v1 = np.sqrt(mu/r1)
v2 = np.sqrt(mu/(r2))
ecc_transfer = (r2-r1)/(r2+r1)
h_transfer = np.sqrt(mu*r1*(1+ecc_transfer))
v_pt = h_transfer/r1
v_at = h_transfer/r2
delV = abs(v_pt-v1) + abs(v_at-v2)
print(delV)
a = (r1+r2)/2
T = 2*np.pi*np.sqrt(a**3/mu)
transfer_time = T/2
print(transfer_time/60)
```
