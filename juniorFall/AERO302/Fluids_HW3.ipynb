Fluid Kinematics 2

```python
# constants
rho = 1.225
g = 9.81
theta = np.deg2rad(5)
h = .005
visc = 1.789e-5

def vort(y):
    return (rho*g*np.sin(theta)*(y-h))/visc

y = np.linspace(0, .005, 1000)

plt.plot(y, vort(y))
plt.xlabel("Lenght of Boundary Layer [m]")
plt.ylabel("Vorticity [Hz]")
plt.show()
```
