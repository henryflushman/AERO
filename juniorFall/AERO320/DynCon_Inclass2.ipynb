```python
from scipy.integrate import solve_ivp
import numpy as np

import src.AERO.AERO320.DynCon_Functions as df


def body_rates_to_quat_dot(q, omega):
    """
    Compute quaternion derivative from body angular rates.

    Parameters
    ----------
    q : array-like, shape (4,)
        Quaternion [q0, q1, q2, q3] (scalar-first convention).
    omega : array-like, shape (3,)
        Body rates [p, q, r] in radians per second.

    Returns
    -------
    q_dot : ndarray, shape (4,)
        Time derivative of the quaternion.
    """
    q0, q1, q2, q3 = q
    p, q_, r = omega

    # Quaternion rate matrix
    Omega = np.array([
        [0,   -p,  -q_,  -r],
        [p,    0,   r,  -q_],
        [q_,  -r,   0,   p],
        [r,    q_, -p,   0]
    ])

    q_dot = 0.5 * Omega @ np.array([q0, q1, q2, q3])
    return q_dot

q0 = df.euler321_to_quat(eulerangs[0], eulerangs[1], eulerangs[2]).T
print(q0)
t0, tf = 0.0, 45.0
t_eval = np.linspace(t0, tf, 1001)

[t, x] = solve_ivp(fun=lambda t, x: body_rates_to_quat_dot(q, omega),
                   t_span=(t0, tf), y0=q0, t_eval=t_eval,
                   rtol=1e-9, atol=1e-12, method='RK45')
```
